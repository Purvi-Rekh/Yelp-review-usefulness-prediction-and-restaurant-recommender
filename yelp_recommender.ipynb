```python
R = data_recommender.pivot_table(index="user_name",columns="name",values="stars")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
type(sim_matrix)
```

```python
sim_df = pd.DataFrame(sim_matrix,columns=R.index,index=R.index)
sim_df.tail(50)
```
